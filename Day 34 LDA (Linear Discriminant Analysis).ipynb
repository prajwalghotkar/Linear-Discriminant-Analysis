```python
dataset=pd.read_csv('Wine.csv')
dataset
```

```python
# Create X Variable - Independent Variables

X=dataset.iloc[:,:-1].values
display(pd.DataFrame(X))
```

```python
# Create Y Variable - Dependent Variables

y=dataset.iloc[:,-1].values
print(y)
```

```python
# Feature Scaling

from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
sc_x=sc.fit_transform(X)
display(pd.DataFrame(sc_x))
```

```python
# PCA with N components= None

from sklearn.decomposition import PCA
pca=PCA()
pca_x=pca.fit_transform(sc_x)
pd.DataFrame(pca_x)
```

```python
# Exaplained Variance Ratio

print(pd.DataFrame(pca.explained_variance_ratio_))
```

```python
# PCA with N components = 2
from sklearn.decomposition import PCA
pca=PCA(n_components=2)
pca_x=pca.fit_transform(sc_x)
display(pd.DataFrame(pca_x))
```

```python
# Exaplined Varianve Ratio

display(pd.DataFrame(pca.explained_variance_ratio_))
```

```python
# Train Test Split

from sklearn.model_selection import train_test_split
pca_x_train,pca_x_test,y_train,y_test=train_test_split(pca_x,y,test_size=0.2,random_state=0)
print(pca_x.shape)
print(pca_x_train.shape)
print(pca_x_test.shape)
```

```python
# Create Logistic Regression

from sklearn.linear_model import LogisticRegression
lr=LogisticRegression(random_state=0)
lr.fit(pca_x_train,y_train)
```

```python
# Prediction

y_pred=lr.predict(pca_x_test)
print(y_pred)
```

```python
# Display Accuracy

from sklearn.metrics import confusion_matrix,accuracy_score
print(accuracy_score(y_pred,y_test))
cm=confusion_matrix(y_pred,y_test)
print(cm)
```
